# ***PCC***

```python
corr = df3.corr()
# Define a list of color values in hexadecimal format


#plt.figure()
plt.figure(figsize=(6, 5)).patch.set_facecolor('white')
plt.xticks(weight='bold')
plt.yticks(weight='bold')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.rcParams['figure.dpi']=1800
sns.heatmap(corr,
            cmap='ocean', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=False, annot_kws={"size": 5.0}, fmt=".2f", square=True);
```

```python
df4 = df3.drop(['No of Pt ', 'Total EA_cls', 'AW-cls', 'Total EN_cls','Most active Pt'], axis=1)
df4
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df4.corr()
#plt.figure()
plt.figure(figsize=(16, 10)).patch.set_facecolor('white')
plt.xticks(weight='bold')
plt.yticks(weight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.rcParams['figure.dpi']=1200
sns.heatmap(corr,
            cmap='ocean', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=False, annot_kws={"size": 5.0}, fmt=".2f", square=True);
```

```python
corr_matrix = df3.corr(method='spearman')
```

```python
from matplotlib import font_manager
from sklearn.preprocessing import MinMaxScaler

# Set high DPI before plotting
plt.rcParams['figure.dpi'] = 500

# Create the figure first
plt.figure(figsize=(11, 7), facecolor='white')

# Draw the heatmap
sns.heatmap(corr_matrix, vmax=1.0, vmin=-1.0, cmap='twilight', annot=False, annot_kws={"size": 5}, fmt='.2f', square=True)

# Style tick labels
plt.xticks(fontsize=10, weight='bold', rotation=45)
plt.yticks(fontsize=10, weight='bold', rotation=0)
plt.tight_layout()
plt.show()
```
