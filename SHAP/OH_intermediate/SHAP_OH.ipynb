```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from math import sqrt
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming x and y are your feature matrix and target variable, respectively
# Replace them with your actual data

# Initialize variables to store the best metrics and corresponding random state
best_random_state = None
min_mae = float('inf')
min_rmse = float('inf')
max_r2 = float('-inf')

# Loop through different random states
for random_state in range(1, 1500):  # You can adjust the range as needed
    # Split the data into training and testing sets
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=random_state)

    # Initialize the Random Forest Regression model
    rfr = RandomForestRegressor()

    # Train the model
    rfr.fit(x_train, y_train)

    # Predictions
    rfr_pred_test = rfr.predict(x_test)
    rfr_pred_train = rfr.predict(x_train)

    # Calculate metrics for test set
    rfr_rmse_test = sqrt(mean_squared_error(y_test, rfr_pred_test))
    rfr_mae_test = mean_absolute_error(y_test, rfr_pred_test)
    rfr_r2_test = r2_score(y_test, rfr_pred_test)

    # Calculate metrics for train set
    rfr_rmse_train = sqrt(mean_squared_error(y_train, rfr_pred_train))
    rfr_mae_train = mean_absolute_error(y_train, rfr_pred_train)
    rfr_r2_train = r2_score(y_train, rfr_pred_train)

    # Update the best metrics and random state if needed
    if rfr_mae_test < min_mae:
        min_mae = rfr_mae_test
        min_rmse = rfr_rmse_test
        max_r2 = rfr_r2_test
        best_random_state = random_state

# Print the best metrics and corresponding random state for RFR
print("Best Random State for RFR:", best_random_state)
print("Min Test RMSE for RFR:", min_rmse)
print("Min Test MAE for RFR:", min_mae)
print("Max Test R2 for RFR:", max_r2)

# Print metrics for training set for RFR
print("Min Train RMSE for RFR:", rfr_rmse_train)
print("Min Train MAE for RFR:", rfr_mae_train)
print("Max Train R2 for RFR:", rfr_r2_train)

# Train the RFR model with the best random state
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=best_random_state)
rfr = RandomForestRegressor()
rfr.fit(x_train, y_train)
rfr_pred_test = rfr.predict(x_test)

# Plotting the graph for RFR
sns.regplot(x=y_train, y=rfr.predict(x_train), color='g', scatter_kws={'s': 70},
            line_kws={'color': 'blue'}, label='Train set')
plt.scatter(y_test, rfr_pred_test, color='r', s=70, label='Test set')

plt.xlabel('Predicted OH Energy', fontsize=12, weight='bold')
plt.ylabel('DFT OH Energy', fontsize=12, weight='bold')
plt.gca().set_aspect('equal', adjustable='box')
plt.xticks(weight='bold')
plt.yticks(weight='bold')
plt.legend(fontsize=16)
plt.show()
```
